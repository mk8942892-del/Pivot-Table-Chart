4. Explain the difference between Sorting and Filtering with an example from this dataset.
- Sorting reorders all data, while filtering dislays a subset of data.

Sorting arranges all records in a specific sequence based on a chosen column's value (e.g.,ascending or descending sales),changing the order of the entire dataset.

Filtering, on the other hand, hides records that do not meet specific criteria (e.g., showing only "Mangors" in the "Product" column), temporarily reducing the visible dataset to a relevant subset.
